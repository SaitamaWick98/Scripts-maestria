```python
# Cargar libreria
df = pd.read_csv('/Users/saitamawick98/Documents/💻/Febrero 2026/Discrepantes/df_tb_bd_inconsistencias_904.csv')
df
```

```python
df_datos = df.loc[:, ["SRA", "Pais", "Drtype", "Linaje", "Sublinaje", "Familia"]]
```

```python
ids = df["SRA"].drop_duplicates().to_list()
ids
```

```python
from pathlib import Path
import pandas as pd

ruta = Path("/Users/saitamawick98/Documents/💻/Febrero 2026/Discrepantes/sift_discrepantes/resultados_sift")

xls_files = [x for x in ruta.iterdir() if x.suffix == ".xls"]

lista_df = []

for xls in xls_files:

    file_id = xls.name.split("_")[0]
    
    if file_id in ids:

        df = pd.read_csv(xls, sep='\t')  # 👈 corregido
        df["Archivo"] = file_id
        df["Tipo"] = "Discrepante"
        lista_df.append(df)

        print(f"{file_id}: Añadido a la lista")
    
    else:
        print(f"{file_id}: No está en ids ❌")
```

```python
df = pd.concat(lista_df, ignore_index=True)
```

```python
df
```

```python
# Buscar bombas de eflujo

# Cargar lista de bombas de eflujo 
bombas_eflujo = Path('/Users/saitamawick98/Documents/💻/Escritorio_antiguo/Tesis Maestria/genes_bombasdeeflujo.txt')

with bombas_eflujo.open('r') as f:
    effluxpump = [bomba.strip() for bomba in f if bomba.strip()]

len(effluxpump)
```

```python
df = df[df["GENE_ID"].isin(effluxpump)]
df
```

```python
df = df.merge(df_datos, left_on="Archivo", right_on="SRA", how="left").drop_duplicates()
```

```python
df.columns
```

```python
df = df[[col for col in df.columns if col not in ["CHROM", "dbSNP", "Archivo"]]]

df.to_csv("/Users/saitamawick98/Documents/💻/Febrero 2026/SIFT_Sus_dis/SIFT_dis_be.csv", index=False)
```

```python
conteo_genes = df[["GENE_ID", "GENE_NAME"]].value_counts(normalize=True, dropna=False).mul(100)
conteo_genes.values
```

```python

labels = [f"{c[1]}" if pd.notna(c[1]) else f"{c[0]}" for c in conteo_genes.index]

plt.figure(figsize=(12,6))
bars = plt.bar(labels, conteo_genes.values)

plt.title("Top genes (proporción)", fontsize=14)
plt.xlabel("Gen")
plt.ylabel("Proporción")

plt.xticks(rotation=60, ha="right")


plt.tight_layout()
plt.show()
```
